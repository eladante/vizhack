```python
import matplotlib.pyplot as plt
from numpy import *
```

```python
plt.figure()
ax = plt.axes()
x = linspace(0,1,100)
ax.plot(x, sin(x))
ax.plot(x,x**3)
```

```python
ax.lines
```

```python
ax.lines.remove(ax.lines[0])
plt.draw()
```

```python
F = ax.fill_between(x,0,x**4)
F
```

```python
ax.collections.remove(F)
plt.draw()
```

```python
T = ax.text(.4,.4,'A')
```

```python
ax.texts.remove(ax.texts[0])
plt.draw()
```

```python
if len(ax.texts) > 0:
    ax.texts.remove(ax.texts[0])
```

```python
plt.figure()
ax = plt.axes()
ax.set_xlim(0,2)
ax.set_ylim(0,2)
plt.plot([1],[1],'ro')
A = ax.annotate('Red dot', xy=(1,1), xycoords='data',
                xytext=(1.3,1.3), 
                va="top", ha="right",
                bbox=dict(boxstyle="round", alpha=0.2),
                arrowprops=dict(
                          arrowstyle="->",
                          connectionstyle="angle,angleA=-95,angleB=35,rad=10"))
```
